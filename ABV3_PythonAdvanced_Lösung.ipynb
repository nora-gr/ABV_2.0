### Aufgabe 4 – Break-Even-Analyse

Im Rahmen einer Unternehmensgründung soll eine Funktion zur Break-Even-Analyse implementiert werden. Der Break-Even-Punkt (BEP) gibt an, ab welcher Absatzmenge ein Unternehmen seine Gesamtkosten durch Erlöse deckt:

$$BEP = \frac{\text{Fixkosten}}{\text{Preis} - \text{variable Stückkosten}}$$

**a)** Schreiben Sie eine Funktion `break_even(fixkosten, preis, variable_kosten)`, die den BEP (gerundet auf ganze Stück) zurückgibt. Geben Sie eine Fehlermeldung aus und geben Sie `None` zurück, falls `preis <= variable_kosten` (kein positiver Deckungsbeitrag).

**b)** Rufen Sie die Funktion mit folgenden Szenarien auf und geben Sie das Ergebnis jeweils als aussagekräftigen Satz aus:

- Fixkosten: 12.000 €, Preis: 8 €, variable Kosten: 8 € *(Fehlerfall)*
- Fixkosten: 80.000 €, Preis: 120 €, variable Kosten: 45 €

<details>
<summary><b>Tipp zu a) – Formel umsetzen</b></summary>

```python
import math

def break_even(fixkosten, preis, variable_kosten):
    if preis <= variable_kosten:
        print("Fehler: Kein positiver Deckungsbeitrag.")
        return None
    bep = fixkosten / (preis - variable_kosten)
    return math.ceil(bep)  # aufrunden auf ganze Stück
```
</details>

<details>
<summary><b>Tipp zu b) – None prüfen</b></summary>

Fangen Sie den Fehlerfall ab, bevor Sie das Ergebnis ausgeben:

```python
    print(f"Break-Even wird bei {bep} Stück erreicht.")
```
</details>

````python
# Ihre Lösung hier:
def break_even(fixkosten, preis, variable_kosten):
    if preis <= variable_kosten:
        return "Der Preis muss höher als die variablen Kosten sein, damit ein Break-Even-Punkt erreicht werden kann."
    
    stueckzahl = fixkosten / (preis - variable_kosten)
    return round(stueckzahl, 2)

test_scenario = [
    {
        "fixkosten": 50000,
        "preis": 25,
        "variable_kosten": 10
    },
    {
        "fixkosten": 12000,
        "preis": 8,
        "variable_kosten": 8
    },
    {
        "fixkosten": 80000,
        "preis": 120,
        "variable_kosten": 45
    }
]

for scenario in test_scenario:
    print(f"Fixkosten: {scenario['fixkosten']} Euro, Preis: {scenario['preis']} Euro, Variable Kosten: {scenario['variable_kosten']} Euro")
    print(f"Break-Even-Stückzahl: {break_even(fixkosten=scenario['fixkosten'], preis=scenario['preis'], variable_kosten=scenario['variable_kosten'])}")
    print(50* "=")
````

````python
# Aufgabe 5a – Klasse implementieren
class Kurs:
    def __init__(self, titel, dozent, max_teilnehmer):
        self.__titel = titel
        self.__dozent = dozent
        self.__max_teilnehmer = max_teilnehmer
        self.__teilnehmer = []
    def anmelden(self, name):
        if len(self.__teilnehmer) < self.__max_teilnehmer:
            self.__teilnehmer.append(name)
            print(f"{name} wurde für den Kurs '{self.__titel}' angemeldet.")
        else:
            print(f"Anmeldung fehlgeschlagen: Der Kurs '{self.__titel}' ist bereits voll.")
    def abmelden(self, name):
        if name in self.__teilnehmer:
            self.__teilnehmer.remove(name)
            print(f"{name} wurde vom Kurs '{self.__titel}' abgemeldet.")
        else:
            print(f"Abmeldung fehlgeschlagen: {name} ist nicht im Kurs '{self.__titel}' angemeldet.")
    def display(self):
        print(f"Der Kurs {self.__titel}, von Dozent*in {self.__dozent} hat aktuell {len(self.__teilnehmer)} von {self.__max_teilnehmer}")

# Testcode (erst ausführen, wenn die Klasse fertig ist):
bwl_grundlagen = Kurs("BWL Grundlagen", "Prof. Dr. Richter", 3)
bwl_grundlagen.anmelden("Anna Bauer")
bwl_grundlagen.anmelden("Ben Schulz")
bwl_grundlagen.anmelden("Clara Hoffmann")
bwl_grundlagen.anmelden("Daniel Meier")  # -> Fehlermeldung erwartet
bwl_grundlagen.abmelden("Ben Schulz")
bwl_grundlagen.abmelden("Felix Wagner")  # -> Fehlermeldung erwartet
bwl_grundlagen.display()

bwl_grundlagen.__max_teilnehmer
````

Implementieren Sie ein einfaches Kassensystem für einen kleinen Hochschulcafé-Stand.

Das System soll Bestellungen entgegennehmen, Preise berechnen und eine Tagesabrechnung ausgeben.

**a) Funktion `berechne_preis(artikel, menge)`**

Schreiben Sie eine Funktion, die den Gesamtpreis für eine Bestellung berechnet.

| Artikel | Preis |
|---------|-------|
| `"Kaffee"` | 1,50 € |
| `"Tee"` | 1,20 € |

Die Funktion soll `None` zurückgeben und eine Meldung ausgeben, wenn der Artikel unbekannt ist.

**b) Klasse `Tageskasse`**

Erstellen Sie eine Klasse `Tageskasse` mit folgenden Anforderungen:

**Attribute:**
- `umsatz` – Gesamtumsatz des Tages, startet bei `0.0`

**Methoden:**

Verarbeiten Sie die folgende Bestellliste automatisch mit einer `for`-Schleife und rufen Sie danach `tagesabrechnung()` auf:

```python
bestellungen_des_tages = [
```

```python
def bestellen(self, artikel, menge):
    betrag = berechne_preis(artikel, menge)
    if betrag is not None:
```

<details>
<summary><b>Tipp zu c) – Schleife über Liste von Tupeln</b></summary>

Entpacken Sie jedes Tupel direkt in der Schleife:
```python
for artikel, menge in bestellungen_des_tages:
    kasse.bestellen(artikel, menge)
```
</details>

````python
bestellungen_des_tages = [
    ("Kaffee", 3),
    ("Snack", 2),
    ("Tee", 5),
    ("Pizza", 1),   # unbekannter Artikel
    ("Wasser", 4),
    ("Kaffee", 7),
]

# Ihre Lösung hier:
def berechne_preis(artikel, menge):
    if artikel == "Kaffee":
        return 1.5 * menge
    elif artikel == "Tee":
        return 1.2 * menge
    elif artikel == "Snack":
        return 2.0 * menge
    elif artikel == "Wasser":
        return 0.8 * menge
    else:
        print(f"Artikel '{artikel}' ist unbekannt. Preis kann nicht berechnet werden.")
        return None
    
class Tageskasse:
    def __init__(self):
        self.umsatz = 0
        self.bestellungen = []
    def bestellen(self, artikel, menge):
        betrag = berechne_preis(artikel, menge)
        if betrag is not None:
            self.umsatz += betrag
            self.bestellungen.append({"artikel": artikel, "menge": menge, "betrag": betrag})
    def tagesabrechnung(self):
        print("Tagesabrechnung:")
        print(50* "=")
        for bestellung in self.bestellungen:
            print(f"{bestellung['menge']}x {bestellung['artikel']} - {bestellung['betrag']:.2f} Euro")
        print(50* "=")
        print(f"Gesamtumsatz: {self.umsatz:.2f} Euro")

tageskasse = Tageskasse()
for artikel, menge in bestellungen_des_tages:
    tageskasse.bestellen(artikel, menge)
tageskasse.tagesabrechnung()
````
